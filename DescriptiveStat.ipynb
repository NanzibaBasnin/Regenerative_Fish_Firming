```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load your dataset
dataset_path = 'ambient_data_05_16_22_metadata.csv'  # Replace this with the path to your dataset
df = pd.read_csv(dataset_path)

# Extract the average production costs
prod_cost_avg_northatlantic = df['prod_cost_avg_northatlantic'].values[0]
prod_cost_avg_southatlantic = df['prod_cost_avg_southatlantic'].values[0]
prod_cost_avg_northpacific = df['prod_cost_avg_northpacific'].values[0]
prod_cost_avg_centralpacific = df['prod_cost_avg_centralpacific'].values[0]
prod_cost_avg_southpacific = df['prod_cost_avg_southpacific'].values[0]
prod_cost_avg_indian = df['prod_cost_avg_indian'].values[0]

oceans = ['North Atlantic', 'South Atlantic', 'North Pacific', 'Central Pacific', 'South Pacific', 'Indian']
avg_costs = [prod_cost_avg_northatlantic, prod_cost_avg_southatlantic, prod_cost_avg_northpacific, prod_cost_avg_centralpacific, prod_cost_avg_southpacific, prod_cost_avg_indian]

def plot_bar_chart(oceans, avg_costs):
    x_pos = np.arange(len(oceans))

    plt.bar(x_pos, avg_costs, align='center', alpha=0.5)
    plt.xticks(x_pos, oceans, rotation=45)
    plt.ylabel('Average Production Cost')
    plt.title('Average Production Cost by Ocean')

    plt.show()

plot_bar_chart(oceans, avg_costs)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset
dataset_path = 'flux_data_07_21_22_metadata.csv'  # Replace this with the path to your dataset
df = pd.read_csv(dataset_path)

# Select the relevant columns
sink_cost_columns = [
    'net_costperton_sink_avg_northatlantic',
    'net_costperton_sink_avg_southatlantic',
    'net_costperton_sink_avg_northpacific',
    'net_costperton_sink_avg_centralpacific',
    'net_costperton_sink_avg_southpacific',
    'net_costperton_sink_avg_indian'
]

sink_cost_data = df[sink_cost_columns]

# Create box plots
plt.figure(figsize=(12, 6))
plt.boxplot(sink_cost_data.values, vert=False, labels=sink_cost_columns)
plt.xlabel('Net Cost per Ton')
plt.title('Box Plots for Net Cost per Ton of Sink Average in Different Ocean Regions')
plt.show()

# Create histograms
fig, axs = plt.subplots(3, 2, figsize=(12, 12))
axs = axs.ravel()

for i, column in enumerate(sink_cost_columns):
    axs[i].hist(sink_cost_data[column], bins=50, alpha=0.75, edgecolor='black')
    axs[i].set_xlabel('Net Cost per Ton')
    axs[i].set_ylabel('Frequency')
    axs[i].set_title(column)

fig.suptitle('Histograms for Net Cost per Ton of Sink Average in Different Ocean Regions')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
dataset_path = 'ambient_data_05_16_22_metadata.csv'  # Replace this with the path to your dataset
df = pd.read_csv(dataset_path)

# Select the relevant columns
product_cost_columns = [
    'net_costperton_product_avg_northatlantic',
    'net_costperton_product_avg_southatlantic',
    'net_costperton_product_avg_northpacific',
    'net_costperton_product_avg_centralpacific',
    'net_costperton_product_avg_southpacific',
    'net_costperton_product_avg_indian'
]

product_cost_data = df[product_cost_columns]

# Calculate skewness and kurtosis
skewness = product_cost_data.skew()
kurtosis = product_cost_data.kurt()

# Combine the skewness and kurtosis into a single DataFrame
stats_df = pd.concat([skewness.rename('skewness'), kurtosis.rename('kurtosis')], axis=1)

# Print the DataFrame as a table
print("Skewness and Kurtosis Table:")
print(stats_df.to_string())

# Visualize the histograms
sns.set_style("whitegrid")
for column in product_cost_columns:
    plt.figure()
    sns.histplot(data=product_cost_data, x=column, kde=True, color='blue', bins=20)
    plt.title(f'Histogram for {column}')
    plt.xlabel('Cost per Tonne')
    plt.ylabel('Frequency')
    plt.show()
```
